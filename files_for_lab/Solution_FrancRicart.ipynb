## 6. Does any column contain alpha and numeric data?
Decide how to clean it

```python
#categorical["Renew Offer Type"] = categorical["Renew Offer Type"].str.strip("Offer")
#categorical.head()

def clean_rot(x):
    y = ""
    for char in x:
        if re.match("[A-Z]|[a-z]", char):
            pass
        else:
            y = y+char
    return y
```

```python
categorical["Renew Offer Type"] = categorical["Renew Offer Type"].apply(clean_rot)
categorical["Renew Offer Type"].value_counts()
```

- We could group the 'employment_status' by 'Employed' and 'other'

```python
def clean_es(x):
        if x == "Employed":
            return x
        else:
            return "Other" 
```

```python
categorical["EmploymentStatus"] = categorical["EmploymentStatus"].apply(clean_es)
categorical["EmploymentStatus"].value_counts()
```

- We could group the luxury vehicles together

```python
categorical["Vehicle Class"] = np.where(categorical["Vehicle Class"].isin(["Luxury Car", "Sports Car", "Luxury SUV"]), "Luxury", categorical["Vehicle Class"])
categorical["Vehicle Class"].value_counts()
```
